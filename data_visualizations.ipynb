## Data Visualizations
Data Visualizations Examples

```python
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.interpolate import make_interp_spline
```

```python
# Set up a clean plotting style and inline rendering for Jupyter
%matplotlib inline
plt.style.use("seaborn-v0_8-whitegrid" if "seaborn-v0_8-whitegrid" in plt.style.available else "default")
```

```python
# Create a common time array for the data
dates = pd.date_range(start="2026-01-01", periods=10, freq="D")
x_days = np.arange(10)  # Numeric representation for spline calculation
```

### Capture a Trend

```python
# -------------------------------------------------------------
# 1. Line Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
y1 = [100, 120, 115, 140, 135, 170, 110, 150, 130, 155]

plt.plot(dates, y1, marker="o", color="#1f77b4", linewidth=2)
plt.title("1. Line Chart (Revenue Over Time)", fontsize=14, fontweight="bold")
plt.xlabel("Date")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 2. Multi-line Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
y2_apple = [150, 155, 152, 165, 160, 175, 170, 185, 180, 190]
y2_amazon = [130, 128, 135, 132, 140, 138, 145, 142, 150, 148]

plt.plot(dates, y2_apple, label="Apple Stock", color="#2ca02c", linewidth=2)
plt.plot(
    dates,
    y2_amazon,
    label="Amazon Stock",
    color="#ff7f0e",
    linewidth=2,
    linestyle="--",
)
plt.title(
    "2. Multi-line Chart (Apple vs Amazon Stock)", fontsize=14, fontweight="bold"
)
plt.xlabel("Date")
plt.ylabel("Stock Price ($)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 3. Area Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
y3 = [50, 65, 55, 85, 80, 100, 45, 75, 60, 80]

plt.plot(dates, y3, color="#ffc107", linewidth=2)
plt.fill_between(dates, y3, color="#ffc107", alpha=0.4)
plt.title("3. Area Chart (Total Sales Over Time)", fontsize=14, fontweight="bold")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 4. Stacked Area Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
# Breakdown subgroups
y4_organic = [30, 40, 35, 50, 45, 60, 25, 45, 35, 50]
y4_paid = [20, 25, 20, 35, 35, 40, 20, 30, 25, 30]

plt.stackplot(
    dates,
    y4_organic,
    y4_paid,
    labels=["Organic Traffic", "Paid Traffic"],
    colors=["#1a2530", "#ffc107"],
    alpha=0.9,
)
plt.title(
    "4. Stacked Area Chart (Active Users by Segment)",
    fontsize=14,
     fontweight="bold",
)
plt.xlabel("Date")
plt.ylabel("Active Users")
plt.legend(loc="upper left")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 5. Spline Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
y5 = [20, 45, 30, 55, 40, 70, 50, 65, 55, 80]

# Interpolate to create smooth curve points
x_smooth = np.linspace(x_days.min(), x_days.max(), 200)
spl = make_interp_spline(x_days, y5, k=3)
y_smooth = spl(x_smooth)

# Map the smoothed x back to a dummy timestamp timeline for proper display
date_ticks = pd.date_range(start="2026-01-01", periods=200, freq="5T")[
    :200
]  # Placeholder for smooth mapping
# Alternative cleaner approach: plot using numeric x and replace ticks labels
plt.plot(x_smooth, y_smooth, color="#757575", linewidth=2, label="Spline Curve")
plt.scatter(x_days, y5, color="#1a2530", zorder=5, label="Actual Data Points")

plt.title(
    "5. Spline Chart (Electricity Consumption)", fontsize=14, fontweight="bold"
)
plt.xlabel("Time Period")
plt.ylabel("Consumption (kWh)")
plt.xticks(x_days, [d.strftime("%b %d") for d in dates], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Set up clean plotting styling
%matplotlib inline
colors_palette = ["#1a2530", "#ffc107", "#d1d5db"]
```

```python
# -------------------------------------------------------------
# 1. Pie Chart
# -------------------------------------------------------------
plt.figure(figsize=(6, 6))
labels_pie = ["Cloud Provider A", "Cloud Provider B", "Cloud Provider C"]
sizes_pie = [45, 30, 25]

plt.pie(
    sizes_pie,
    labels=labels_pie,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors_palette,
)
plt.title("1. Pie Chart (Market Share of Cloud Providers)", fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 2. Donut Pie Chart
# -------------------------------------------------------------
plt.figure(figsize=(6, 6))
labels_donut = ["Android", "iOS", "Others"]
sizes_donut = [70, 25, 5]

# Plot standard pie chart
plt.pie(
    sizes_donut,
    labels=labels_donut,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors_palette,
    pctdistance=0.85,
)

# Draw a white circle in the center to turn it into a donut
centre_circle = plt.Circle((0, 0), 0.70, fc="white")
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("2. Donut Pie Chart (Android OS Market Share)", fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 3. Heat Map
# -------------------------------------------------------------
plt.figure(figsize=(8, 5))
# Generate synthetic monthly temperature data across 4 zones
np.random.seed(42)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
zones = ["Zone 1", "Zone 2", "Zone 3", "Zone 4"]
data_heatmap = np.random.randint(15, 38, size=(4, 12))

sns.heatmap(
    data_heatmap,
    annot=True,
    xticklabels=months,
    yticklabels=zones,
    cmap="YlGnBu_r",
    cbar_kws={"label": "Temperature (°C)"},
)
plt.title("3. Heat Map (Average Monthly Temperatures)", fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# 4. Stacked Column Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 5))
regions = ["North", "East", "South", "West"]
q1_sales = [40, 55, 30, 45]
q2_sales = [35, 60, 40, 50]
q3_sales = [50, 45, 35, 55]

# Plot bars sequentially on top of each other
plt.bar(regions, q1_sales, label="Q1", color="#1a2530")
plt.bar(regions, q2_sales, bottom=q1_sales, label="Q2", color="#ffc107")
plt.bar(
    regions,
    q3_sales,
    bottom=np.array(q1_sales) + np.array(q2_sales),
    label="Q3",
    color="#d1d5db",
)

plt.title("4. Stacked Column Chart (Quarterly Sales per Region)", fontweight="bold")
plt.xlabel("Regions")
plt.ylabel("Sales (Units)")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 5. Treemap Chart
# -------------------------------------------------------------
plt.figure(figsize=(9, 5))
categories = [
    "Produce\n(£45k)",
    "Dairy\n(£430k)",
    "Meat\n(£420k)",
    "Bakery\n(£415k)",
    "Frozen\n(£410k)",
]
sizes_treemap = [45, 30, 20, 15, 10]
# Use custom shades matching the diagram colors
treemap_colors = ["#ffc107", "#ffd54f", "#1a2530", "#2c3e50", "#d1d5db"]

squarify.plot(
    sizes=sizes_treemap, label=categories, color=treemap_colors, alpha=0.9
)
plt.title("5. Treemap Chart (Grocery Sales Count with Categories)", fontweight="bold")
plt.axis("off")  # Turn off the canvas axes
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 1. Card Component
# -------------------------------------------------------------

fig, ax = plt.subplots(figsize=(5, 3))
ax.set_facecolor("#f8f9fa")
fig.patch.set_facecolor("#ffffff")

# Draw a stylized frame or boundary card block
rect = plt.Rectangle(
    (0.05, 0.05),
    0.9,
    0.9,
    transform=ax.transAxes,
    color="#1a2530",
    alpha=0.05,
    zorder=1,
)
ax.add_patch(rect)

# Add KPI Labels and Value
plt.text(
    0.5,
    0.65,
    "$7.47M",
    fontsize=36,
    fontweight="bold",
    color="#1a2530",
    ha="center",
    va="center",
)
plt.text(
    0.5,
    0.35,
    "Total Revenue to Date",
    fontsize=12,
    color="#6c757d",
    ha="center",
    va="center",
)
plt.text(
    0.5,
    0.18,
    "▲ +12.4% vs last month",
    fontsize=10,
    color="green",
    ha="center",
    va="center",
)

ax.axis("off")
plt.title("1. KPI Card", fontsize=14, fontweight="bold", pad=10)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 2. Table Chart
# -------------------------------------------------------------
fig, ax = plt.subplots(figsize=(7, 3))
ax.axis("off")

# Mock leaderboard dataset
data_table = {
    "Account Executive": ["Alice Smith", "Bob Jones", "Charlie Brown"],
    "Registrations": [142, 115, 98],
    "Target Achieved": ["118%", "96%", "82%"],
}
df = pd.DataFrame(data_table)

# Render table layout
table_obj = ax.table(
    cellText=df.values,
    colLabels=df.columns,
    cellLoc="center",
    loc="center",
    colColours=["#1a2530", "#1a2530", "#1a2530"],
)

# Disable auto-sizing so our manual scale/font changes take effect
table_obj.auto_set_font_size(False) 

# Option A: Use the internal font size setter method
table_obj.set_fontsize(11) 

# Alternatively, Option B (safer across older versions):
# Loop through the cells to set the size directly:
# for cell in table_obj.values():
#     cell.set_fontsize(11)

table_obj.scale(1.2, 2)  # Stretch cell scale dimensions padding

# Style the headers specifically text-wise
for col_idx in range(len(df.columns)):
    cell = table_obj[0, col_idx]
    cell.get_text().set_color("white")
    cell.get_text().set_weight("bold")

plt.title(
    "2. Table Chart (Account Executive Leaderboard)",
    fontsize=14,
    fontweight="bold",
    pad=20,
)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 3. Gauge Chart (NPS Score Metric Indicator)
# -------------------------------------------------------------
# Setting target gauge value: Let's track an NPS score of 65 on a scale of 0-100
nps_score = 65
arrow_angle = 180 - (nps_score / 100.0 * 180)  # Maps value directly to degrees

# Create polar coordinate subplot
fig, ax = plt.subplots(figsize=(6, 4), subplot_kw={"projection": "polar"})

# Gauge rings tracking limits (Poor, Good, Excellent)
boundaries = [0, 60, 140, 180]
colors_gauge = ["#d1d5db", "#ffc107", "#1a2530"]  # Match image palette layout

for i in range(len(colors_gauge)):
    ax.bar(
        x=np.deg2rad((boundaries[i] + boundaries[i + 1]) / 2),
        height=0.4,
        width=np.deg2rad(boundaries[i + 1] - boundaries[i]),
        bottom=0.6,
        color=colors_gauge[i],
        align="center",
    )

# Pass explicit numerical width and head properties directly to the dictionary
ax.annotate(
    "",
    xy=(np.deg2rad(arrow_angle), 1.0),  # End pointer point tip
    xytext=(0, 0),                      # Central base origin
    arrowprops=dict(
        facecolor="black",
        edgecolor="black",
        width=2,         # Width of the arrow tail stem
        headwidth=8,     # Total width of the arrow head base
        headlength=10,   # Total length of the arrow head
        shrink=0.05
    ),
)

# Clean polar gauge display settings
ax.set_theta_zero_location("E")
ax.set_xlim(0, np.pi)  # Show only top semi-circle axis span
ax.set_thetagrids([0, 72, 144, 180], labels=["100", "60", "20", "0"])
ax.set_yticklabels([])
ax.grid(False)

# Add explicit center text display readout indicator
plt.text(
    0,
    0,
    f"NPS: {nps_score}",
    fontsize=16,
    fontweight="bold",
    ha="center",
    va="center",
)
plt.title(
    "3. Gauge Chart (Executive KPI Dashboard Target)",
    fontsize=14,
    fontweight="bold",
    y=1.1,
)
plt.tight_layout()
plt.show()
```

```python
# Generate a base random normal dataset for realistic distribution visualizations
np.random.seed(42)
base_data = np.random.normal(loc=70000, scale=15000, size=500)  # e.g., Salary distribution
group_a = np.random.normal(loc=24, scale=4, size=100)           # e.g., Group 1 efficiency
group_b = np.random.normal(loc=28, scale=5, size=100)           # e.g., Group 2 efficiency
group_c = np.random.normal(loc=20, scale=3, size=100)           # e.g., Group 3 efficiency
```

```python
# -------------------------------------------------------------
# 1. Histogram
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
plt.hist(base_data, bins=10, color="#1a2530", edgecolor="white", rwidth=0.95)

plt.title("1. Histogram (Distribution of Salaries)", fontsize=14, fontweight="bold")
plt.xlabel("Salary ($)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 2. Box Plot
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
box_data = [group_a, group_b, group_c]

# Create boxplot with custom color filling layout
bp = plt.boxplot(box_data, patch_artist=True, labels=["Vehicle Class A", "Vehicle Class B", "Vehicle Class C"])

colors_box = ["#1a2530", "#ffc107", "#757575"]
for patch, color in zip(bp["boxes"], colors_box):
    patch.set_facecolor(color)
    patch.set_alpha(0.9)

# Style median lines to remain clearly visible
for median in bp["medians"]:
    median.set(color="white" if median.get_xdata()[0] < 2 else "black", linewidth=2)

plt.title("2. Box Plot (Gas Efficiency of Vehicles)", fontsize=14, fontweight="bold")
plt.xlabel("Vehicle Category")
plt.ylabel("Miles Per Gallon (MPG)")
plt.tight_layout()
plt.show()
```
