### Capture a Trend

```python
# -------------------------------------------------------------
# 1. Line Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
y1 = [100, 120, 115, 140, 135, 170, 110, 150, 130, 155]

plt.plot(dates, y1, marker="o", color="#1f77b4", linewidth=2)
plt.title("1. Line Chart (Revenue Over Time)", fontsize=14, fontweight="bold")
plt.xlabel("Date")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 2. Multi-line Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
y2_apple = [150, 155, 152, 165, 160, 175, 170, 185, 180, 190]
y2_amazon = [130, 128, 135, 132, 140, 138, 145, 142, 150, 148]

plt.plot(dates, y2_apple, label="Apple Stock", color="#2ca02c", linewidth=2)
plt.plot(
    dates,
    y2_amazon,
    label="Amazon Stock",
    color="#ff7f0e",
    linewidth=2,
    linestyle="--",
)
plt.title(
    "2. Multi-line Chart (Apple vs Amazon Stock)", fontsize=14, fontweight="bold"
)
plt.xlabel("Date")
plt.ylabel("Stock Price ($)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 3. Area Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
y3 = [50, 65, 55, 85, 80, 100, 45, 75, 60, 80]

plt.plot(dates, y3, color="#ffc107", linewidth=2)
plt.fill_between(dates, y3, color="#ffc107", alpha=0.4)
plt.title("3. Area Chart (Total Sales Over Time)", fontsize=14, fontweight="bold")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 4. Stacked Area Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
# Breakdown subgroups
y4_organic = [30, 40, 35, 50, 45, 60, 25, 45, 35, 50]
y4_paid = [20, 25, 20, 35, 35, 40, 20, 30, 25, 30]

plt.stackplot(
    dates,
    y4_organic,
    y4_paid,
    labels=["Organic Traffic", "Paid Traffic"],
    colors=["#1a2530", "#ffc107"],
    alpha=0.9,
)
plt.title(
    "4. Stacked Area Chart (Active Users by Segment)",
    fontsize=14,
     fontweight="bold",
)
plt.xlabel("Date")
plt.ylabel("Active Users")
plt.legend(loc="upper left")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 5. Spline Chart
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
y5 = [20, 45, 30, 55, 40, 70, 50, 65, 55, 80]

# Interpolate to create smooth curve points
x_smooth = np.linspace(x_days.min(), x_days.max(), 200)
spl = make_interp_spline(x_days, y5, k=3)
y_smooth = spl(x_smooth)

# Map the smoothed x back to a dummy timestamp timeline for proper display
date_ticks = pd.date_range(start="2026-01-01", periods=200, freq="5T")[
    :200
]  # Placeholder for smooth mapping
# Alternative cleaner approach: plot using numeric x and replace ticks labels
plt.plot(x_smooth, y_smooth, color="#757575", linewidth=2, label="Spline Curve")
plt.scatter(x_days, y5, color="#1a2530", zorder=5, label="Actual Data Points")

plt.title(
    "5. Spline Chart (Electricity Consumption)", fontsize=14, fontweight="bold"
)
plt.xlabel("Time Period")
plt.ylabel("Consumption (kWh)")
plt.xticks(x_days, [d.strftime("%b %d") for d in dates], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 3. Gauge Chart (NPS Score Metric Indicator)
# -------------------------------------------------------------
# Setting target gauge value: Let's track an NPS score of 65 on a scale of 0-100
nps_score = 65
arrow_angle = 180 - (nps_score / 100.0 * 180)  # Maps value directly to degrees

# Create polar coordinate subplot
fig, ax = plt.subplots(figsize=(6, 4), subplot_kw={"projection": "polar"})

# Gauge rings tracking limits (Poor, Good, Excellent)
boundaries = [0, 60, 140, 180]
colors_gauge = ["#d1d5db", "#ffc107", "#1a2530"]  # Match image palette layout

for i in range(len(colors_gauge)):
    ax.bar(
        x=np.deg2rad((boundaries[i] + boundaries[i + 1]) / 2),
        height=0.4,
        width=np.deg2rad(boundaries[i + 1] - boundaries[i]),
        bottom=0.6,
        color=colors_gauge[i],
        align="center",
    )

# Pass explicit numerical width and head properties directly to the dictionary
ax.annotate(
    "",
    xy=(np.deg2rad(arrow_angle), 1.0),  # End pointer point tip
    xytext=(0, 0),                      # Central base origin
    arrowprops=dict(
        facecolor="black",
        edgecolor="black",
        width=2,         # Width of the arrow tail stem
        headwidth=8,     # Total width of the arrow head base
        headlength=10,   # Total length of the arrow head
        shrink=0.05
    ),
)

# Clean polar gauge display settings
ax.set_theta_zero_location("E")
ax.set_xlim(0, np.pi)  # Show only top semi-circle axis span
ax.set_thetagrids([0, 72, 144, 180], labels=["100", "60", "20", "0"])
ax.set_yticklabels([])
ax.grid(False)

# Add explicit center text display readout indicator
plt.text(
    0,
    0,
    f"NPS: {nps_score}",
    fontsize=16,
    fontweight="bold",
    ha="center",
    va="center",
)
plt.title(
    "3. Gauge Chart (Executive KPI Dashboard Target)",
    fontsize=14,
    fontweight="bold",
    y=1.1,
)
plt.tight_layout()
plt.show()
```

```python
# Generate a base random normal dataset for realistic distribution visualizations
np.random.seed(42)
base_data = np.random.normal(loc=70000, scale=15000, size=500)  # e.g., Salary distribution
group_a = np.random.normal(loc=24, scale=4, size=100)           # e.g., Group 1 efficiency
group_b = np.random.normal(loc=28, scale=5, size=100)           # e.g., Group 2 efficiency
group_c = np.random.normal(loc=20, scale=3, size=100)           # e.g., Group 3 efficiency
```

```python
# -------------------------------------------------------------
# 1. Histogram
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
plt.hist(base_data, bins=10, color="#1a2530", edgecolor="white", rwidth=0.95)

plt.title("1. Histogram (Distribution of Salaries)", fontsize=14, fontweight="bold")
plt.xlabel("Salary ($)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 2. Box Plot
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
box_data = [group_a, group_b, group_c]

# Create boxplot with custom color filling layout
bp = plt.boxplot(box_data, patch_artist=True, labels=["Vehicle Class A", "Vehicle Class B", "Vehicle Class C"])

colors_box = ["#1a2530", "#ffc107", "#757575"]
for patch, color in zip(bp["boxes"], colors_box):
    patch.set_facecolor(color)
    patch.set_alpha(0.9)

# Style median lines to remain clearly visible
for median in bp["medians"]:
    median.set(color="white" if median.get_xdata()[0] < 2 else "black", linewidth=2)

plt.title("2. Box Plot (Gas Efficiency of Vehicles)", fontsize=14, fontweight="bold")
plt.xlabel("Vehicle Category")
plt.ylabel("Miles Per Gallon (MPG)")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 3. Violin Plot
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))
# Combine data into a structured DataFrame for Seaborn mapping
df_violin = pd.DataFrame({
    "Time Spent (min)": np.concatenate([group_a, group_b, group_c]),
    "Age Group": ["18-25"]*100 + ["26-45"]*100 + ["46+"]*100
})

sns.violinplot(
    x="Age Group", 
    y="Time Spent (min)", 
    data=df_violin, 
    palette=["#ffc107", "#d1d5db", "#1a2530"]
)

plt.title("3. Violin Plot (Time Spent in Restaurants across Age Groups)", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 4. Density Plot (KDE)
# -------------------------------------------------------------
plt.figure(figsize=(8, 4))

# Plot two overlapping segments for comparative visualization
sns.kdeplot(base_data, fill=True, color="#1a2530", alpha=0.9, label="Segment Premium")
sns.kdeplot(base_data - 15000, fill=True, color="#ffc107", alpha=0.6, label="Segment Economy")

plt.title("4. Density Plot (Comparing Price Distribution by Segment)", fontsize=14, fontweight="bold")
plt.xlabel("Price ($)")
plt.ylabel("Density")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# 5. Bubble Chart
# -------------------------------------------------------------

# *Use Case: Relationship between Life Expectancy (Y), GDP per capita (X), & Population size (Bubble Size)*

# Generate Mock Data
num_countries = 30
gdp_per_capita = np.random.uniform(5000, 55000, num_countries)
# Modest positive correlation between GDP and Life Expectancy
life_expectancy = 60 + (gdp_per_capita / 2000) + np.random.normal(0, 3, num_countries)
population_millions = np.random.uniform(10, 500, num_countries)

# Plot
plt.figure(figsize=(9, 5.5))
# Scaling bubble sizes for clarity
bubble_sizes = population_millions * 3 

scatter = plt.scatter(gdp_per_capita, life_expectancy, s=bubble_sizes, 
                      c='#1d2a44', alpha=0.7, edgecolors='black')

plt.xlabel('GDP per Capita ($)')
plt.ylabel('Life Expectancy (Years)')
plt.title('Life Expectancy vs. GDP per Capita (Bubble Size = Population)')

# Optional: Add a legend showing bubble size mapping
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.5, num=4)
labels = ['100M', '200M', '300M', '400M'] # Representative clean labels
plt.legend(handles, labels, loc="lower right", title="Population")

plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import plotly.graph_objects as go
```

```python
# -------------------------------------------------------------
# 1. Sankey Chart
# -------------------------------------------------------------

# *Use Case: Supply chain volumes between warehouses*

# Define nodes (warehouses and destinations)
# Note: Plotly Sankey requires integer indices for source and target
nodes = ["Warehouse A", "Warehouse B", "Hub North", "Hub South", "Customer 1", "Customer 2"]

# Define the flows (Source index -> Target index) with volumes (value)
sources = [0, 0, 1, 1, 2, 2, 3, 3]
targets = [2, 3, 2, 3, 4, 5, 4, 5]
values  = [40, 20, 15, 45, 30, 25, 20, 45]

# Plot using Plotly (perfect for Sankey flows)
fig_sankey = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = nodes,
      color = "#1d2a44"
    ),
    link = dict(
      source = sources,
      target = targets,
      value = values,
      color = "rgba(29, 42, 68, 0.2)" # Semi-transparent dark blue/gray
    )
)]) 

fig_sankey.update_layout(title_text="Sankey Chart: Supply Chain Volumes Between Warehouses", font_size=12)
fig_sankey.show()
```

```python
# -------------------------------------------------------------
# 2. Chord Chart (Circular Flow Abstraction)
# -------------------------------------------------------------

# *Use Case: Export between countries to showcase biggest export partners*
# *Note: Since standard chord charts require external Javascript engines (like d3.js), an interactive circular flow map using Plotly is used here to best replicate the image logic.*

# Define matrix mapping exports between US, China, and Germany
countries = ["US", "China", "Germany", "Japan"]
# Matrix: Row exports to Column
matrix = [
    [0, 50, 30, 20],  # US exports
    [65, 0, 25, 40],  # China exports
    [40, 35, 0, 15],  # Germany exports
    [30, 25, 10, 0]   # Japan exports
]

# Flattening matrix into coordinates on a circle to create the "chord loop" structure
num_nodes = len(countries)
angles = np.linspace(0, 2 * np.pi, num_nodes, endpoint=False)
x_coords = np.cos(angles)
y_coords = np.sin(angles)

fig_chord = go.Figure()

# Draw the links between regions
for i in range(num_nodes):
    for j in range(num_nodes):
        if matrix[i][j] > 0:
            fig_chord.add_trace(go.Scatter(
                x=[x_coords[i], x_coords[j]],
                y=[y_coords[i], y_coords[j]],
                mode='lines',
                line=dict(width=matrix[i][j]*0.15, color='rgba(230, 175, 46, 0.6)' if i==1 else 'rgba(29, 42, 68, 0.4)'),
                hoverinfo='text',
                text=f"{countries[i]} to {countries[j]}: ${matrix[i][j]}B",
                showlegend=False
            ))

# Draw outer node points
fig_chord.add_trace(go.Scatter(
    x=x_coords, y=y_coords, mode='markers+text',
    marker=dict(size=25, color='#1d2a44'),
    text=countries, textposition="top center",
    showlegend=False
))

fig_chord.update_layout(
    title="Chord/Circular Relationship Chart: Country Export Volumes",
    xaxis=dict(visible=False), yaxis=dict(visible=False),
    width=600, height=600, plot_bgcolor='rgba(0,0,0,0)'
)
fig_chord.show()
```
